### Задание 1 и 2
- Составить реестр, в который войдут все клиенты из отчета 1
- Недостающие поля взять из отчета 2

```python
reestr = pd.merge(left=report_1,
                    right=report_2,
                    left_on='Номер контракта',
                    right_on='NumContract',
                    how='left')
reestr.drop(columns=['NumContract', 'LoanAmount'])

reestr.head()
```

### Задание 3
- Из сформированного реестра исключить клиентов из вкладки "цессия" и со статусом "банкрот", "погашен", "досрочно погашен"

```python
reestr = reestr[
    (~reestr['ContractID'].isin(cessiya['ContractID'])) &
    (~reestr['Status'].isin(['EARLY_REPAID', 'REPAID', 'BANKRUPTCY']))]

reestr.head()
```

### Задание 4
- Из полученного реестра построить сводную таблицу по регионам, количеством клиентов по регионам и суммой выданных займов по регионам

```python
pivot_reestr = reestr.pivot_table(
    index='SubdivisionName',
    values=['Номер контракта', 'LoanAmount'],
    aggfunc={'Номер контракта':'count', 'LoanAmount': 'sum'}
).rename(columns={'LoanAmount': 'total_amount', 'Номер контракта':'count'})

pivot_reestr.head()
```

## Часть 2

Подготовка

```python
import sqlite3

conn = sqlite3.connect(':memory:')

report_2.to_sql('report_2', conn, index=False, if_exists='replace')
report_1.to_sql('report_1', conn, index=False, if_exists='replace')

def query(sql_query):
    return pd.read_sql_query(sql_query, conn)
```

### Задание 1
- Написать SQL запрос по отчету 2, который выведет клиентов с регионом выдачи онлайн и выданные в 2024 году


```python
query('''
SELECT ClientID
FROM report_2
WHERE SubdivisionName = 'Онлайн'
      AND strftime('%Y', [issueDateTimestamp]) = '2024'
''')
```

### Задание 2
- Написать SQL запрос по отчету 2, который выведет три поля: статус, количество клиентов в статусе, сумму выданных займов в статусе

```python
query('''
SELECT Status, COUNT(ClientID) as clients, SUM(LoanAmount) as total_amount
FROM report_2
GROUP BY Status
ORDER BY COUNT(ClientID) DESC
''')
```

### Задание 3
- Написать SQL запрос, который выведет все поля отчета 1 и отчета 2 (использую LEFT JOIN и подтягиваю данные к отчет 2, если нужны все то использую RIGHT)

```python
query('''
SELECT *
FROM report_2 r2
LEFT JOIN report_1 r1
    ON r2.ContractID = r1.[Номер контракта]
LIMIT 5
''')
```
